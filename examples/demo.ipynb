# ONNX Inference Demo
This notebook demonstrates running the exported ONNX model on a sample image using the project's inference utilities.

Steps:
1. Create a temporary sample image.
2. Call `predict_image()` from `src.api_inference`.
3. Inspect the returned output summary.

```python
from src.api_inference import predict_image
from PIL import Image
import tempfile, os

# Create a temporary RGB image
img = Image.new('RGB', (300, 200), color=(120, 150, 200))
fd, path = tempfile.mkstemp(suffix='.jpg')
os.close(fd)
try:
    img.save(path, format='JPEG')
    print('Running prediction on', path)
    result = predict_image('models/onnx/best.onnx', path)
    print(result)
finally:
    os.remove(path)
```
